Wgraj zawartość pliku orders.csv do ramki danych. Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/drive/MyDrive/LAB04_pandas cz2/orders.csv', sep=',')
print(data.head())
print(data.info())
print(data.describe())
```

Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;

```python
data['order_date'] = pd.to_datetime(data['order_date'])
data
```

Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;

```python
print(data['tshirt_category'].unique)
print(data['tshirt_category'].nunique)
```

Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

```python
data['tshirt_category'].replace({'Wh Tshirt M':'White T-Shirt M', 
                                 'Wh Tshirt F':'White T-Shirt F', 
                                 'Bl Tshirt F': 'Black T-Shirt F',
                                 'Bl Tshirt M':'Black T-Shirt M', 
                                 'Tennis Shirt':'Other Tennis-Shirt M/F',
                                 'Hoodie':'Other Hoodie M/F'},
                                inplace = True)
print(data['tshirt_category'].unique())
print(data['tshirt_category'].nunique())
```

Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
data[['tshirt_color', 'tshirt_type', 'tshirt_sex']] = data['tshirt_category'].str.split(expand=True)
data
```

Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016

```python
daterange = (data['order_date'] >= '2014-12-31') & (data['order_date'] <= '2016-08-16')

data = data.loc[daterange]
data
```
